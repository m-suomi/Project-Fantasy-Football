```python
import pandas as pd
```

```python
over_unders = pd.read_csv('week_3_over_unders.csv')
over_unders
```

```python
high = over_unders['over_under'].max()
print(high)
low = over_unders['over_under'].min()
print(low)
mean = over_unders['over_under'].mean()
print(mean)
```

```python
(high+mean+mean)/3
```

```python
#create bins from range of over unders
bins = [(low-.5), ((low+low+mean)/3), ((low+mean+mean)/3), ((high+mean+mean)/3), ((high+high+mean)/3), (high+.5)]
over_unders['binned'] = pd.cut(over_unders['over_under'], bins)
over_unders.head()
```

```python
over_unders['binned'].value_counts()
```

```python
#convert binned values to string so we can categorize with adjectives
over_unders['binned'] = over_unders['binned'].map(lambda x: str(x))
```

```python
over_unders['binned'].iloc[0]
```

```python
#create adjectives from bins
score_bins = {
    '(37.0, 40.146]': 'Terrible',
    '(40.146, 42.792]': 'Concerned',
    '(42.792, 48.958]': 'Neutral',
    '(48.958, 52.479]': 'Good',
    '(52.479, 56.5]': 'Awesome'
}
```

```python
over_unders['score_sentiment'] = over_unders['binned'].map(score_bins)
over_unders.head()
```

```python
over_unders.to_csv('week_3_score_sent.csv')
```
